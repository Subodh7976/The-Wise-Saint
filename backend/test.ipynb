```python
from tools import MathsToolSpec
from llama_index.tools.wikipedia import WikipediaToolSpec
from llama_index.core.agent import ReActAgent
from llama_index.llms.gemini import Gemini


llm = Gemini(model_name="models/gemini-pro", temperature=0)
tool_list = WikipediaToolSpec().to_tool_list()
tool_list.extend(MathsToolSpec().to_tool_list())

agent = ReActAgent.from_tools(tool_list, llm=llm, verbose=True, context="Use each tool and multiple times if required to fully explain the given user query.")
```

```python
agent.query("Plot 4x - 3y = 2, and 2y + 3x = 7 and solve the equation ")
```
